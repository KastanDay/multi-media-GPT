````python
raw_dataset = tf.data.TFRecordDataset(filenames[2])
for raw_record in raw_dataset.take(1):
  # IMPORTANT: use SequenceExample (for frame level) vs Example (for segment level)
  example = tf.train.SequenceExample() # tf.train.Example() 
  example.ParseFromString(raw_record.numpy())
  raw_record.numpy()
  print(example)
````
